```python
import pandas as pd
```

```python
data = {
    'Id':['S01','S02','S03','S04','S05'],
    'Student':['Amit','Steve','John','Virat','Rohit'],
    'Rank':[1,4,3,5,2],
    'Marks':[95,70,80,60,90]

}
```

```python
df = pd.DataFrame(data)
```

```python
for _ in df.iterrows():
    print(_, '\n')
```

```python
for _ in df.itertuples():
    print(_, '\n')
```

```python
for _ in df.columns:
    print(_, '\n')
```

```python
for a, b in df.items():
    print(a, b)
```
