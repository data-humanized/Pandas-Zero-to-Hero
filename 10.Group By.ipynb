```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Id':['S01','S02','S03','S04','S05'],
    'Student':['Rohit','Steve','Virat','Virat','Rohit'],
    'Rank':[1,4,3,5,2],
    'Marks':[95,70,80,60,90]

}
```

```python
df = pd.DataFrame(data)
```

```python
#To print the first occurance of each entry
first_non_null_entries = df.groupby('Student').first()
first_non_null_entries
```

### Iterating the Group


```python
grp = df.groupby('Student')
```

```python
for name, group in grp:
    print(name,'\n',group,'\n')
```

### View the group using Groups method

```python
df
```

```python
grp.groups
```

### applying aggregations on the groups

```python
mean = df.groupby('Student')['Marks'].mean()
mean
```

```python
players = {
    'Player':["Amit","John","Amit","David","Kane","John"],
    'Points':[95,98,92,99,94,91],
    "Year":[2011,2015,2011,2019,2015,2019]
}
```

```python
player_df = pd.DataFrame(data=players)
```

##### Creating Groups


```python
points_per_year = player_df.groupby(['Year','Player'])
```

##### Aggregation on the groups

```python
points_per_year['Points'].sum()
```

### Using Numpy with Grouby

```python
points_per_year['Points'].agg(np.mean)
```

### Preview the size of the Groups

```python
points_per_year['Points'].agg(np.size)
```
