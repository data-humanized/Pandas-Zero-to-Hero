# Joining Pandas Dataframs

```python
data1 = {
    "id":["S01","S02","S03","S04","S05"],
    'Student':['Amit','John','Jacob','David','Steve'],
    'Roll':[101, 102, 103, 104, 105]
}

data2 = {
    'Rank':[1,4,3,5],
    'Marks':[95,70,80,60]
}
```

```python
dataFrame1 = pd.DataFrame(data1)
dataFrame2 = pd.DataFrame(data2)

print("dataFrame1 =\n",dataFrame1)
print("\n\n")
print("dataFrame2 =\n",dataFrame2)
```

```python
# Joining the dataframe

resultant = dataFrame1.join(dataFrame2) # the dataframe 1 will have all the values but data frame 2 will loose or miss values as per the first dataframe
```

```python
resultant
```

# Concatenating dataframes

```python
concat1 = {
    "id":["S01","S02","S03","S04","S05"],
    'Student':['Amit','John','Jacob','David','Steve'],
    'Roll':[101, 102, 103, 104, 105]
}
```

```python
concat2 = {
    'id':['S06','S07','S08'],
    'Student':['Benjamin','Isabella','Jacob'],
}
```

```python
concat1df = pd.DataFrame(concat1, index = ['Student-1', 'Student-2', 'Student-3','Student-4','Student-5'])
concat2df = pd.DataFrame(concat2, index= ['Student-6', 'Student-7', 'Student-8'])
concat_result = pd.concat([concat1df, concat2df], ignore_index=True)
```

```python
concat_result
```

```python
concat_result.iloc[6,1]
```

# Combining two series


```python
Series1 = pd.Series([10,20,40,80,100])
Series2 = pd.Series([25,5,75,95,45])
```

```python
def demo(x1,x2):
    if (x1 > x2):
        return x1
    else:
        return x2
```

```python
combine_series = Series1.combine(Series2,demo) # Combining means it compares and keeps the value based on the function condition
```

```python
combine_series
```

```python
combine_using_lambda = Series1.combine(Series2, lambda x1,x2: max(x1,x2))
combine_using_lambda
```

# Concatenating Series

```python
Series_concat = pd.concat([Series1,Series2], ignore_index=True)
Series_concat
```
