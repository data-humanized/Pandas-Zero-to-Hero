```python
import pandas as pd
```

```python
print(pd.__version__)
```

```python
data = {
    'Id':['S01','S02','S03','S04','S05'],
    'Student':['Amit','Steve','John','Virat','Rohit'],
    'Rank':[1,4,3,5,2],
    'Marks':[95,70,80,60,90],
    'Boolean':[True, False, True, False, True]

}
```

```python
df = pd.DataFrame(data)
```

```python
df.dtypes
```

```python
print(df)
```

```python
df[['Student', 'Id']] = df[['Student', 'Id']].apply(lambda x:x.str.lower())
```

```python
df = df.apply(lambda col: col.str.upper() if col.dtype == 'str' else col)
```

```python
print(df)
```

```python
df = df.apply(lambda col: col.str.strip() if col.dtype == 'str' else col)
```

```python
df['Student'].apply(repr)
```

```python
df.apply(lambda col: col.str.contains('ROHIT') if col.dtype == 'str' else col)
```
