### Adding new columns in dataframe

```python
data = {
    'Id':['S01','S02','S03','S04','S05'],
    'Student':['Amit','Steve','John','Virat','Rohit'],
    'Rank':[1,4,3,5,2],
    'Marks':[95,70,80,60,90]

}
```

```python
DF  = pd.DataFrame(data)
```

```python
DF.insert(1,"Roll_Number",[101,102,103,104,105])
DF
```

```python
DF.drop(columns = ["Roll_Number"], inplace=True,errors="ignore")
```

### AddinUsing assign function

```python
DF = DF.assign(Roll = ['1','2','3','4','5'])
DF['ROLLNEW'] = ['10','20','30','40','50']
```

```python
DF = DF.assign(Delete = ['d','d','d','d','d'])
```

```python
DF
```

#### Drop a column

    three ways of removing a column
    Using .drop(columns = ['Col1','Col2])
    Using .drop(['ColName','ColName2'], asis = column
    Uding .drop(['ColName','ColName2'], axis = 1)  # 1 stands for axis

```python
DF.drop(columns = ['ROLLNEW'],inplace = True,errors="ignore")
```

```python
DF.drop(['Roll'], inplace = True,errors="ignore", axis = 'columns')
```

```python
DF.drop('Delete', inplace = True, errors="ignore", axis = 1)
```

```python
DF
```

### To remove an index, i.e a row
    using axis = 0 or axis = 'index'

```python
DF.drop([1,2],errors = "ignore", inplace = True, axis = 0)
```

```python
DF.drop(0, errors = "ignore", axis = 'index', inplace = True)
DF
```
